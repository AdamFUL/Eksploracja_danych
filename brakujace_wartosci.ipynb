## **Uzupełnienie średnią lub modą​**

W pandas *dataFrame* możemy uzupełnić brakujące wartości za pomocą funkcji `fillna()`.
* Przyjmuje ona jedną wartość, która będzie wstawiana w pustych pozycjach lub zamiast wartości NaN.

Uzupełnia wszystkie brakujące wartości w kolumnie age średnią obliczoną z tejże kolumny

```data['age'] = data.age.fillna(data.age.mean())​```

Uzupełnia wszystkie brakujące wartości w kolumnie income medianą obliczoną z tejże kolumny

```data['income']=data.income.fillna(data.income.median())```

Zastępuje wszystkie brakujące wartości w kolumnie gender (kolumna kategorii) dominantą wyliczoną z tejże kolumny​

```data['gender']=data['gender'].fillna(data['gender'].mode()[0])​```

## **Obsługa brakujących danych jako stringi**

**Problem**
* Dane często zawierają braki zapisane jako tekst: **"NA", "N/A", "brak", "?", "-", "brak danych"**

* Pandas **nie rozpoznaje** ich automatycznie jako brakujące wartości!


### **Rozwiązanie 1: Wczytywanie z pliku CSV**

```
import pandas as pd

# Definiujemy listę wartości oznaczających brak
braki = ['NA', 'N/A', 'brak', 'brak danych', '?', '-', '']

# Wczytujemy plik z parametrem na_values
df = pd.read_csv('dane.csv', na_values=braki)

```

### **Rozwiązanie 2: Zamiana po wczytaniu**

```
# Zamieniamy wybrane stringi na NaN
df.replace(['NA', 'N/A', 'brak', 'brak danych', '?', '-'], np.nan, inplace=True)

# LUB dla konkretnej kolumny
df['kolumna'] = df['kolumna'].replace('brak', np.nan)
```

Wszystkie dane niezgodne z parametrami tratuję jako NaN, jako iż w przypadku mieszkań istnieje wysoka możliwość iż mieszkanie o któym brakuje nam danych jest poza spektrum naszej mediany innych mieszkań.

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

na_vals = ["brak", "Brak", "BRAK", "?", "-999", "none", "None", "NULL", "null", "n/a", "NA", "nan", ""]
df = pd.read_csv("nieruchomosci.csv", na_values=na_vals, keep_default_na=True)

num_cols = ['powierzchnia', 'rok_budowy', 'pietro', 'liczba_pieter', 'cena_m2', 'cena_total']
for col in num_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')

if 'powierzchnia' in df.columns:
    df.loc[df['powierzchnia'] == 0, 'powierzchnia'] = np.nan

if 'pietro' in df.columns:
    df.loc[df['pietro'] == -999, 'pietro'] = np.nan

if 'balkon' in df.columns:
    df['balkon'] = df['balkon'].astype(str).str.strip()
    df.loc[df['balkon'] == '?', 'balkon'] = np.nan

if 'stan' in df.columns:
    df['stan'] = df['stan'].astype(str).str.strip()
    allowed_stany = {'Dobry', 'Bardzo dobry', 'Do remontu'}
    df.loc[~df['stan'].isin(allowed_stany), 'stan'] = np.nan

for col in ['cena_m2', 'cena_total']:
    if col in df.columns:
        df.loc[df[col] < 0, col] = np.nan

print("\nLiczba brakujących wartości po ignorowaniu niezgodnych danych:")
missing = df.isna().sum()
print(missing)

filtered_df = df.dropna(subset=['powierzchnia', 'cena_total'])

plt.figure(figsize=(8,6))
plt.scatter(filtered_df['powierzchnia'], filtered_df['cena_total'], alpha=0.7)
plt.title('Cena do powierzchni')
plt.xlabel('Powierzchnia mieszkania [m²]')
plt.ylabel('Cena całkowita [PLN]')
plt.grid(True, linestyle='--', alpha=0.6)

if not filtered_df.empty:
    z = np.polyfit(filtered_df['powierzchnia'], filtered_df['cena_total'], 1)
    p = np.poly1d(z)
    plt.plot(filtered_df['powierzchnia'], p(filtered_df['powierzchnia']), "r--", label='trend liniowy')
    plt.legend()

plt.show()
````
